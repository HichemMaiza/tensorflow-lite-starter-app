# Convert from .h5 keras model to .tflite model

## First part -- train and save the keras model 


```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.contrib import lite
```

```python
# define the model, one layer and one unit
model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])]) 
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
#get some data 
xs =np.array([i*1.0 for i in range(200)] ) 
ys =np.array([i*2.0 for i in xs])
model.fit(xs,ys, epochs = 700)
```

```python
# test the model 
print(model.predict([10.0]))
```

```python
# save the model 
keras_file = "linear_model.h5"
keras.models.save_model(model, keras_file)
```

## Working attempts using two-steps conversion 

```python
from tensorflow.python.framework import graph_util
from tensorflow.python.framework import graph_io
import tensorflow as tf

net_model = tf.keras.models.load_model("linear_model.h5")

tf.keras.backend.set_learning_phase(0)
sess = tf.keras.backend.get_session() 

pred = [None]*1
pred_node_names = [None]*1

for i in range(1):
    pred_node_names[i] = 'output_node'+str(i)
    pred[i] = tf.identity(net_model.outputs[i], name=pred_node_names[i])

print('output nodes names are: ', pred_node_names)

constant_graph = graph_util.convert_variables_to_constants(sess, sess.graph.as_graph_def(), pred_node_names)
graph_io.write_graph(constant_graph, '.' , "newModel.pb", as_text=False)
```
